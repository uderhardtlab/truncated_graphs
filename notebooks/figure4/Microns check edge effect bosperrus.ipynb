# Exploring the edge effect in the data

For this exploration, we need the full representation of the connectome, so we load it in "conntility" format.
This package can be installed using:
 > pip install Connectome-Utilities
 
The other packages are standard

```python
import pandas
import numpy
import conntility
from matplotlib import pyplot as plt
import bosperrus
```

```python
def identify_safe_neurons_bosperrus(node_df):
    flow = bosperrus.Flow.from_coords_and_scores(coordinates=node_df[["x_nm", "z_nm"]], scores=node_df[["indegree", "outdegree"]], distance_fn=bosperrus.distance_to_convex_hull)
    flow.flow()
    safe = flow.observations.distance_to_convex_hull > flow.best_fits["indegree"].params['piecewise_linear_b']
    return safe, flow.best_fits["indegree"].params['piecewise_linear_b']
```

```python
fn_mat = "/data/bionets/je30bery/bosperrus-experiments/microns_data/microns_mm3_connectome.h5"
name_dset_f = "full"
name_dset_c = "condensed"

M = conntility.ConnectivityMatrix.from_h5(fn_mat, name_dset_f)
node_df = M.vertices

_, elbow = identify_safe_neurons_bosperrus(node_df)
print(elbow)
```
