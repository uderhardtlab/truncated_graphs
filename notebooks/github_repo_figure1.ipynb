```python
import sys
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
sys.path.append("../../bosporus-package/bosporus/")
from distances import *
```

```python
test_coords = np.array([[0, 1], [0, 2], [0, 3], [0, 4], [1, 0], [1, 1], [1, 2], [1, 3], [2, 1], [2, 2], [2, 3], [2, 4]])
```

```python
mask = np.zeros([3, 3], dtype=bool)
mask[1,:2] = 1
```

```python
d1 = distance_to_rectangular_border(test_coords)
pointset = np.array([[1.5, 1.5]])
d2 = distance_to_pointset(test_coords, pointset=pointset)
d3 = distance_to_mask(test_coords, mask)
d4 = distance_to_convex_hull(test_coords)
```

```python
f, axs = plt.subplots(1, 4, figsize=(10, 2))
axs[0].scatter(test_coords[:, 0], test_coords[:, 1], c=d1, cmap="plasma")
axs[0].plot([0, 0], [0, 4], c="red")
axs[0].plot([2, 2], [0, 4], c="red")
axs[0].plot([0, 2], [0, 0], c="red")
axs[0].plot([0, 2], [4, 4], c="red")
axs[0].set_title("...rectangular image border")
axs[1].scatter(test_coords[:, 0], test_coords[:, 1], c=d2, cmap="plasma")
axs[1].scatter(pointset[:, 0], pointset[:, 1], c="red", s=20)
axs[1].set_title("...point set")
axs[2].scatter(np.where(mask)[0], np.where(mask)[1], c="red", s=500)
axs[2].scatter(test_coords[:, 0], test_coords[:, 1], c=d3, cmap="plasma")
axs[2].set_title("...mask")
axs[3].scatter(test_coords[:, 0], test_coords[:, 1], c=d4, cmap="plasma")
axs[3].plot([0, 0], [1, 4], c="red")
axs[3].plot([2, 2], [1, 4], c="red")
axs[3].plot([0, 1], [1, 0], c="red")
axs[3].plot([1, 2], [0, 1], c="red")
axs[3].plot([0, 2], [4, 4], c="red")
axs[3].set_title("...convex hull")
for ax in axs:
    ax.axis("off")
    ax.set_xlim(-0.5, 2.5)
    ax.set_ylim(-0.5, 4.5)
    ax.set_aspect(1)
f.suptitle("Nodes colored by distance to...")
plt.tight_layout()
plt.savefig("../../bosporus-package/plots_readme/distance_functions.svg", bbox_inches="tight")
```
